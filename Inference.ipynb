```python
from detect.yolo.inference import run, run_per_img
from models.common import DetectMultiBackend
import torch
```

# Fed with live videos

```python
import cv2
data = "./data/custom.yaml"
weights = "./runs/train/yolov5n/weights/best.pt"
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model = DetectMultiBackend(weights, device=device, data=data)

source = "./data/dataset/Fall_Detection_Object/Test/Fall_Person_valid_000000.png"
img = cv2.imread(source)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (416, 416))


view_img = True

frame = run_per_img(model, img, device, view_img=view_img)
```
